```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
```

3 Transaction 01- Write data as delta format

```python
df.write.format("delta").save("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta")
```

3 Transaction 02- Overwrite delta format

```python
from pyspark.sql.functions import col
df.filter(col("country") == "India").write.format("delta").save("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta",
                              mode="overwrite")
```

#  maintain versioning

```python
spark.read.option("versionAsOf", 1).load("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta").display()
```
