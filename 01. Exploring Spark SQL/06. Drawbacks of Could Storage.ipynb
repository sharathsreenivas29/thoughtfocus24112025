# Prerequisites

## Load users_001.csv into Dataframe

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
```

# Writing into parquet format

```python
#df_parquest = df.write.format("parquet").save("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquest")
df_parquest = spark.read.parquet("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquest")
```

Create view

```python
df_parquest = df.write.format("parquet").save("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquest")
df_parquest.createOrReplaceTempView("users_vw")
```
